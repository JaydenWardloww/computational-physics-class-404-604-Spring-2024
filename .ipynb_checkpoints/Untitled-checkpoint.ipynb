```python
# Homework

Generate a new jupyter notebook file for this homework

Write a program to plot
\begin{equation}
f(x)=\pi/2+2*(sin(x)+sin(3x)/3+sin(5x)/5+....sin(nx)/n)
\end{equation}
from -$\pi$ to $\pi$. Choose n=11, 21 and 31 and plot all these three curves in one figure. Please use numpy library and matplotlib.

Then commit this to your github.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calc(x, n):
    function = np.pi/2
    for i in range(1, n+1, 2): #this will only be giving me integers in odd intervals 
        function += 2*np.sin(i*x)/i
    return function

#this is from -np.pi to np.pi, I thought i would need to use np.arange but I guess not 
x1 = np.linspace (-np.pi, np.pi, 10000) 
n1 = calc(x1, 11)
n2 = calc(x1, 21) 
n3 = calc(x1, 31)
plt.plot(x1, n1, "r" ,label = "n = 11")
plt.plot(x1, n2, "g" ,label = "n = 21")
plt.plot(x1, n3, "b" ,label = "n = 31")
plt.title('Plot of F(x)')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend(loc = "upper left")
plt.grid(True)
plt.show()

        
    



        
        
        
    




```
