```python
import matplotlib.pyplot as plt
import numpy as np

#initial conditions
g = 9.81 #gravity
tend = 3
t = 0
y = [0., 10.]
h = 0.1
#our equation vector
def f(t, y):
    f=np.zeros(2)
    f[0] = y[1]
    f[1] = -g*t
    return f
    
def rk4(tend, h, y, n):
    k1 = np.zeros(n)
    k2 = np.zeros(n)
    k3 = np.zeros(n)
    k4 = np.zeros(n)
    k1 = h*f(t,y)
    k2 = h*f(t+0.5*h,y+0.5*k1)
    k3 = h*f(t+0.5*h,y+0.5*k2)
    k4 = h*f(t+h,y+k3)
    y = y + (k1+2*k2+2*k3+k4)/3
    return y

tp=[]
yp=[]

while(t<tend):
    if((t+h)>tend):
        h=tend-t
    y = rk4(t,h,y,2)
    yp=np.append(yp,y[0])
    t = t+h
    tp=np.append(tp,t)

plt.plot(tp,yp)
plt.show()




```
